The data set I have selected contains caffeine content of a variety of drinks. The data set contains five fields.
<ul>
    <li> drink - specifying the drink's name
    <li> volume (ml) - specifying the drink's total volume in millileters
    <li> Calories - the drink's total calories

    <li> type - the category the drink falls in to

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df_caffeine = pd.read_csv("Data\caffeine.csv")
```

## 2. Identify a question or question(s) that you would like to explore in your data set.

What drink types are the most potentent in terms of caffeine content?

## 3. Create at least three graphs that help answer these questions. Make sure your graphs are clearly readable and are labeled appropriately and professionally.

```python
# Create a scatter plot of caffeine content vs volume
fig=plt.figure()
ax=fig.add_axes([0, 0, 1, 1])
ax.scatter(df_caffeine['Caffeine (mg)'], df_caffeine['Volume (ml)'], color='r')
ax.set_xlabel('Volume (ml)')
ax.set_ylabel('Caffeine Content (mg)')
ax.set_title('Caffeine Content vs Volume')
plt.show()
```

In this first chart I wanted to investigate if volume greatly impacted caffeine content. Based on the chart, I do not think that volume is the main driver. 

```python
# Find the different types of drinks contained in the data set
df_caffeine.type.unique()
```

```python
# Seperate out the caffeine content by drink type
coffee = df_caffeine.loc[df_caffeine['type'] == 'Coffee','Caffeine (mg)']
energy_drink = df_caffeine.loc[df_caffeine['type'] == 'Energy Drinks','Caffeine (mg)']
energy_shots = df_caffeine.loc[df_caffeine['type'] == 'Energy Shots','Caffeine (mg)']
soft_drinks = df_caffeine.loc[df_caffeine['type'] == 'Soft Drinks','Caffeine (mg)']
tea = df_caffeine.loc[df_caffeine['type'] == 'Tea','Caffeine (mg)']
water = df_caffeine.loc[df_caffeine['type'] == 'Water','Caffeine (mg)']

# Seperate out volume by drink type
coffee_v = df_caffeine.loc[df_caffeine['type'] == 'Coffee','Volume (ml)']
energy_drink_v = df_caffeine.loc[df_caffeine['type'] == 'Energy Drinks','Volume (ml)']
energy_shots_v = df_caffeine.loc[df_caffeine['type'] == 'Energy Shots','Volume (ml)']
soft_drinks_v = df_caffeine.loc[df_caffeine['type'] == 'Soft Drinks','Volume (ml)']
tea_v = df_caffeine.loc[df_caffeine['type'] == 'Tea','Volume (ml)']
water_v = df_caffeine.loc[df_caffeine['type'] == 'Water','Volume (ml)']
```

```python
# Create Box Plot
fix, ax = plt.subplots()
c_bp = ax.boxplot(coffee, positions=[2], widths = 1.5, patch_artist=True,
                showmeans=False, showfliers=False, medianprops={"color": "white", "linewidth": 0.5},
                boxprops={"facecolor": "C5", "edgecolor": "white",
                          "linewidth": 0.5},
                whiskerprops={"color": "C5", "linewidth": 1.5},
                capprops={"color": "C5", "linewidth": 1.5})
ed_bp = ax.boxplot(energy_drink, positions=[4], widths = 1.5, patch_artist=True,
                showmeans=False, showfliers=False, medianprops={"color": "white", "linewidth": 0.5},
                boxprops={"facecolor": "C4", "edgecolor": "white",
                          "linewidth": 0.5},
                whiskerprops={"color": "C4", "linewidth": 1.5},
                capprops={"color": "C4", "linewidth": 1.5})
es_bp = ax.boxplot(energy_shots, positions=[6], widths = 1.5, patch_artist=True,
                showmeans=False, showfliers=False, medianprops={"color": "white", "linewidth": 0.5},
                boxprops={"facecolor": "C1", "edgecolor": "white",
                          "linewidth": 0.5},
                whiskerprops={"color": "C1", "linewidth": 1.5},
                capprops={"color": "C1", "linewidth": 1.5})
sd_bp = ax.boxplot(soft_drinks, positions=[8], widths = 1.5, patch_artist=True,
                showmeans=False, showfliers=False, medianprops={"color": "white", "linewidth": 0.5},
                boxprops={"facecolor": "C2", "edgecolor": "white",
                          "linewidth": 0.5},
                whiskerprops={"color": "C2", "linewidth": 1.5},
                capprops={"color": "C2", "linewidth": 1.5})
t_bp = ax.boxplot(tea, positions=[10], widths = 1.5, patch_artist=True,
                showmeans=False, showfliers=False, medianprops={"color": "white", "linewidth": 0.5},
                boxprops={"facecolor": "C3", "edgecolor": "white",
                          "linewidth": 0.5},
                whiskerprops={"color": "C3", "linewidth": 1.5},
                capprops={"color": "C3", "linewidth": 1.5})
w_bp = ax.boxplot(water, positions=[12], widths = 1.5, patch_artist=True,
                showmeans=False, showfliers=False, medianprops={"color": "white", "linewidth": 0.5},
                boxprops={"facecolor": "C0", "edgecolor": "white",
                          "linewidth": 0.5},
                whiskerprops={"color": "C0", "linewidth": 1.5},
                capprops={"color": "C0", "linewidth": 1.5})
ax.axes.xaxis.set_visible(False)
ax.axes.set_ylabel('Caffeine Content (mg)')
ax.axes.set_title('Caffeine Content by Drink Type')
ax.legend([c_bp["boxes"][0],ed_bp["boxes"][0], es_bp["boxes"][0], sd_bp["boxes"][0], t_bp["boxes"][0], w_bp["boxes"][0]], 
          ["Coffee", "Energy Drink", "Energy Shot", "Soft Drink", "Tea", "Water"], loc='upper right')
plt.show
```

This next chart displays the distribution of caffeine content based on drink type. From the chart, the three most potent drink types are coffee, energy drink, and energy shot. Of the three, energy shots tend to be the most potent based on their overall distribution. 

```python
# Create a bar chart of maximum caffeine content
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
Type = ["Coffee", "Energy Drink", "Energy Shot", "Soft Drink", "Tea", "Water"]
Max_CC = [max(coffee), max(energy_drink), max(energy_shots), max(soft_drinks), max(tea), max(water)]
ax.bar(Type, Max_CC)
ax.axes.set_ylabel('Max Caffeine Content (mg)')
ax.axes.set_title('Maximum Caffeine Content by Drink Type')
plt.show()
```

This final chart displays the maximum caffeine content of each drink type. This chart shows that individual coffee drinks contain the most caffeine.

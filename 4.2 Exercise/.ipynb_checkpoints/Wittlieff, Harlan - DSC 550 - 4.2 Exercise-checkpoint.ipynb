## 1. Load the data as a Pandas data frame and ensure that it imported correctly.

```python
# Import the pandas library
import pandas as pd

# Load the datafile
df = pd.read_csv(r'Data/auto-mpg.csv')

# Validate the data loaded correctly
df.head()
```

## 2. Begin by prepping the data for modeling:

### Remove the car name column.

```python
# Drop the column
df = df.drop('car name', 1)

# Validate the drop worked correctly
df.head()
```

### The horsepower column values likely imported as a string data type. Figure out why and replace any strings with the column mean.

```python
# Obtain the counts of each value in the hosepower column
df.groupby('horsepower').count()
```

Question marks were entered into the data where horsepower is uknown

```python
# Replace the ? with NaN
df = df.apply(pd.to_numeric, errors='coerce')

# Show NaN values for horsepower
df[df['horsepower'].isna()]
```

```python
# Calculate the mean horsepower
hp_mean = df['horsepower'].mean()

# Replace NaN with the mean
df['horsepower'] = df['horsepower'].fillna(hp_mean)

# View row 32 to validate horsepower replacement with mean
df.iloc[[32]]
```

### Create dummy variables for the origin column

```python
# Add dummy variables for origin column to original dataframe
df = pd.concat([df, pd.get_dummies(df['origin'])], axis=1)

# View dataframe
df.head()
```

## 3. Create a correlation coefficient matrix and/or visualization. Are there features highly correlated with mpg?

```python
# Build correlation matrix
corrM = df.corr()

# Display correlation matrix
corrM
```

Cylinders, Displacement, Horsepower, and Weight are highly negatively correlated with mpg.

## 4. Plot mpg versus weight. Analyze this graph and explain how it relates to the corresponding correlation coefficient.

```python
# Import library
import matplotlib.pyplot as plt

# Create a scatter plot of mpv versus weight
fig=plt.figure()
ax=fig.add_axes([0, 0, 1, 1])
ax.scatter(df['weight'], df['mpg'], color='r')
ax.set_xlabel('Weight')
ax.set_ylabel('MPG')
ax.set_title('Weight vs MPG')
plt.show()
```

The graph also displays a negative correlation, as weight increases MPG decreases

## 5. Randomly split the data into 80% training data and 20% test data, where your target is mpg.

```python
# Create x & y arrays
x = df[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 1, 2, 3]]
y = df['mpg']

# Import library
from sklearn.model_selection import train_test_split

# Create training & test datasets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

## 6. Train an ordinary linear regression on the training data.

```python
# Import libraries
from sklearn.linear_model import LinearRegression

# create a model
model = LinearRegression()
model.fit(x_train, y_train)

# View model coefficients
pd.DataFrame(model.coef_, x.columns, columns = ['Coeff'])
```

## 7. Calculate R2, RMSE, and MAE on both the training and test sets and interpret your results.

```python
# Import library
from sklearn import metrics

# Build predictions
test_predictions = model.predict(x_test)
train_predictions = model.predict(x_train)

# Calculate metrics
print('Test Metrics:')
print('R2', metrics.r2_score(y_test, test_predictions))
print('RMSE', metrics.mean_squared_error(y_test, test_predictions, squared=False))
print('MAE', metrics.mean_absolute_error(y_test, test_predictions))
print('\nTrain Metrics:')
print('R2', metrics.r2_score(y_train, train_predictions))
print('RMSE', metrics.mean_squared_error(y_train, train_predictions, squared=False))
print('MAE', metrics.mean_absolute_error(y_train, train_predictions))
```

R2: A little over 80% of the variability in mpg can be explained with our model.
MAE: On average we can expect our model to be ~2.6 mpg off from actual values. 
There are not significant changes to RMSE and MAE from the training dataset to the test dataset meaning our model does have some predictive value.

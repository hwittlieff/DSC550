## 10. Try to calculate the median user score of all the video games. You will likely run into an error because some of the user score entries are a non-numerical string that cannot be converted to a float. Find and replace this string with NaN and then calculate the median. Then, replace all NaN entries in the user score column with the median value.

```python
# Find the different values in the user score column
df_videogames.User_Score.unique()
```

```python
# text of 'tbd' identified, replace with NaN
df_videogames['User_Score'] = df_videogames['User_Score'].replace('tbd', np.nan)

# Convert User_Score to numeric
df_videogames['User_Score'] = pd.to_numeric(df_videogames['User_Score'])
```

```python
# Calculate the mean
user_score_mean = df_videogames['User_Score'].mean()
print('Mean User_Score: ',user_score_mean)
```

```python
# Replace all NaN entries in the user score column with the median value
df_videogames['User_Score'] = df_videogames['User_Score'].replace(np.nan, user_score_mean)
```
